```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('salary.csv')
df
```

```python
inputs = df.drop('slary', axis = 'columns')
inputs
```

```python
target = df.slary
target
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['company_n'] = le_company.fit_transform(inputs['company'])
inputs['job_n'] = le_company.fit_transform(inputs['job'])
inputs['degree_n'] = le_company.fit_transform(inputs['degree'])
```

```python
inputs
```

```python
inputs.drop(['company', 'job', 'degree'], axis = 'columns', inplace=True)
inputs
```

if we do not put inplace=True, the columns do not been deoprd so, we have to store the data frame in another variable


**The Data Frame Without inplace = True looks like this

```python
inputs_n = inputs.drop(['company', 'job', 'degree'], axis = 'columns')
inputs_n
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n, target)
```

```python
model.score(inputs, target) * 100
```

```python
result = model.predict([[2, 1, 1]])
```

```python
if (result[0] != 0) :
    print("Salary Is More Than $100")
else : print("Salary Is less Than $100")
```
