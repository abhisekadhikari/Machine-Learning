test_size = Test Semple

test_size = 0.3 means 30% samples are test samples and remaining 70% are train samples


test_size has 0.25 by default which means it takes 25% test sample and remaining 75% are training samples

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```
