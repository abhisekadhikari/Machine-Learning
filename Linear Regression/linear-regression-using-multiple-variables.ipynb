```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import math
```

```python
df = pd.read_csv("homepriceswithbedrooms.csv")
df
```

```python
df.Bedrooms.median()
```

```python
df.Bedrooms = df.Bedrooms.fillna(df.Bedrooms.median())
df
```

```python
# Creating Model
model = linear_model.LinearRegression()
model.fit(df.drop('Price', axis = 'columns'), df.Price)
```

```python
model.predict([[3000, 3, 40]])
```

```python
model.coef_
```

```python
model.intercept_
```
